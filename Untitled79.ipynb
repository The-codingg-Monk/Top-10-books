```python
ratings_with_name=books.merge(ratings,on="ISBN")
```

```python
intel_users=ratings_with_name.groupby("User-ID").count()[ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200].index
```

```python
books_200_ui=ratings_with_name.loc[ratings_with_name["User-ID"].isin(intel_users)]
```

```python
bokks_50=books_200_ui.groupby("Book-Title").count()[books_200_ui.groupby("Book-Title").count()["Book-Rating"]>=50].index
```

```python
most_popular_books=books_200_ui.loc[books_200_ui["Book-Title"].isin(bokks_50)]
```

```python
most_popular_books
```

```python
pt=most_popular_books.pivot_table(index="Book-Title",columns='User-ID',values="Book-Rating")
pt.fillna(0,inplace=True)
pt.index[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine=cosine_similarity(pt)
```

```python
def recommend(book):
    book_index=np.where(pt.index==book)[0][0]
    book_cosine=cosine[book_index]
    top_5=sorted(list(enumerate(book_cosine)),key=lambda x:x[1],reverse=True)[1:6]
    for ind in top_5:
        print(pt.index[ind[0]])
        
  
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
